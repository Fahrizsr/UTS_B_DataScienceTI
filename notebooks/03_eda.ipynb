```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/iot_data_clean.csv", parse_dates=['timestamp'])
df = df.set_index('timestamp')

df.head()
```

```python
df['listrik'].plot(figsize=(12,5))
plt.title("Konsumsi Listrik")
plt.show()

df.plot(subplots=True, figsize=(12,8))
plt.show()

df.describe()
df['hour'] = df.index.hour

df.groupby('hour')['listrik'].mean().plot()
plt.title("Rata-rata Konsumsi Listrik per Jam")
plt.show()

df[df['listrik'] > df['listrik'].mean() + 3*df['listrik'].std()]

df.corr()
```
