```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/iot_data_raw.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df = df.sort_values('timestamp')
df = df.set_index('timestamp')

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

df = df.resample('5min').mean()

df.to_csv("../data/iot_data_clean.csv")
```
