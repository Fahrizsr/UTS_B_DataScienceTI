```python
import pandas as pd
import numpy as np

# 1. Buat timestamp (tiap 5 menit selama 1 hari)
date_range = pd.date_range(start="2024-01-01", periods=288, freq="5min")

# 2. Buat data normal (simulasi sensor)
np.random.seed(42)

listrik = np.random.normal(loc=150, scale=20, size=len(date_range))
suhu = np.random.normal(loc=27, scale=2, size=len(date_range))
sampah = np.random.normal(loc=50, scale=10, size=len(date_range))

# 3. Gabungkan ke DataFrame
df = pd.DataFrame({
    "timestamp": date_range,
    "listrik": listrik,
    "suhu": suhu,
    "sampah": sampah
})

# 4. Tambahkan missing value
for col in ["listrik", "suhu", "sampah"]:
    df.loc[np.random.choice(df.index, size=20, replace=False), col] = np.nan

# 5. Tambahkan outlier ekstrem
df.loc[np.random.choice(df.index, 5), "listrik"] = 500
df.loc[np.random.choice(df.index, 5), "suhu"] = 100

# 6. Rusak format timestamp (biar sesuai soal)
df.loc[0, "timestamp"] = "01-01-2024 00:00"
df.loc[1, "timestamp"] = "2024/01/01 00:05"

# 7. Simpan ke CSV
df.to_csv("../data/iot_data_raw.csv", index=False)

# 8. Tampilkan data
print(df.head())
```
